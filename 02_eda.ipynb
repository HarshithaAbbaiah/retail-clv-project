```python
# Import libraries for data manipulation and visualization
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned retail dataset and automatically convert InvoiceDate to datetime format
df = pd.read_csv("Data/cleaned_retail.csv", parse_dates=["InvoiceDate"])
```

```python
# Calculate total revenue by country, sort in descending order, and display the top 10 countries by revenue
top_countries = df.groupby("Country")["Revenue"].sum().sort_values(ascending=False).head(10)
print(top_countries)
```

```python
# Create a bar chart to visualize the top 10 countries by total revenue
top_countries.plot(kind="bar", figsize=(12,5), color="steelblue")

# Add a title to the chart
plt.title("Top 10 Countries by Revenue")

# Label the y-axis
plt.ylabel("Revenue")

# Rotate country names for better readability
plt.xticks(rotation=45)

# Adjust layout to prevent label overlap
plt.tight_layout()

# Display the chart
plt.show()
```

```python
# Extract the year-month from InvoiceDate and create a Month column
df["Month"] = df["InvoiceDate"].dt.to_period("M")

# Calculate total revenue for each month
monthly_revenue = df.groupby("Month")["Revenue"].sum()

# Plot the monthly revenue trend
monthly_revenue.plot(figsize=(12,5), color="green")

# Add a title to the chart
plt.title("Monthly Revenue Trend")

# Label the y-axis
plt.ylabel("Revenue")

# Adjust layout for better visualization
plt.tight_layout()

# Display the chart
plt.show()
```

```python
# Calculate total quantity sold for each product, rank them in descending order, and display the top 10 best-selling products
top_products = df.groupby("Description")["Quantity"].sum().sort_values(ascending=False).head(10)
print(top_products)
```

```python
# Filter the dataset to include only transactions from customers in the United Kingdom
df_uk = df[df["Country"] == "United Kingdom"]

# Display the dimensions of the UK-only dataset
print(df_uk.shape)
```
