 **Load the Dataset**
 

```python
# Import the Pandas library for data manipulation and analysis
import pandas as pd
```

```python
# Load the online retail dataset from a CSV file using Latin-1 encoding
# Use Latin-1 (ISO-8859-1) encoding to correctly read non-UTF-8 characters in the dataset
df = pd.read_csv("Data/online_retail_II.csv", encoding="latin-1")
```

```python
df.head()# Display the first 5 rows of the DataFrame
```

```python
# Get the number of rows and columns in the DataFrame
df.shape
```

```python
# Display a summary of the DataFrame, including column names, data types, non-null counts, and memory usage
df.info()
```

**Clean the Data**

```python
'''Customer ID has missing values
1067371 - 824364 = 243,007 rows have no customer ID. 
We can't do customer analysis without knowing who bought so we'll drop these rows.'''
df = df.dropna(subset = ["Customer ID"])
print(df.shape)
```

```python
# Keep only records with a positive quantity and display the updated DataFrame dimensions
df = df[df["Quantity"] > 0]
print(df.shape)
```

```python
# Transform InvoiceDate from a string/object to a datetime data type for time-based analysis
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])
print(df["InvoiceDate"].dtype)
```

```python
# Create a Revenue column by multiplying Quantity and Price, then preview the first few rows of the DataFrame
df["Revenue"] = df["Quantity"] * df["Price"]
df.head()
```

```python
# Save the cleaned DataFrame to a CSV file without including the row index, then confirm the save operation
df.to_csv("Data/cleaned_retail.csv", index=False)
print("Saved!")
```
